**Q1. Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.**

Seaborn offers a rich array of plot types for effective data visualization. Here are five commonly used plots and their applications:

Line Plot:

Connects data points with straight lines, suitable for visualizing trends over time or continuous data.
Use cases: Time series analysis, showing changes in values over time.
Scatter Plot:

Displays relationships between two numerical variables, useful for identifying patterns and correlations.
Use cases: Correlation analysis, outlier detection, clustering.
Bar Plot:

Compares categorical data using rectangular bars, representing counts or summary statistics.
Use cases: Comparing values across different categories, visualizing frequencies.
Box Plot:

Visualizes the distribution of numerical data, showing median, quartiles, and outliers.
Use cases: Understanding data spread, identifying outliers, comparing distributions.
Histogram:

Displays the distribution of numerical data by grouping values into bins.
Use cases: Understanding data distribution, identifying patterns and shapes.

**Q2. Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x = "timepoint" and y = "signal" for different events and regions.**

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the fmri dataset
fmri = sns.load_dataset("fmri")

# Create a line plot for different events and regions
sns.lineplot(x="timepoint", y="signal", hue="event", style="region", data=fmri)

plt.show()
```

**Q3. Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x = 'pclass', y = 'age' and y = 'fare'**

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the titanic dataset
titanic = sns.load_dataset("titanic")

# Create two box plots
sns.boxplot(x="pclass", y="age", data=titanic)
plt.show()

sns.boxplot(x="pclass", y="fare", data=titanic)
plt.show()
```

**Q4. Use the "diamonds" dataset from seaborn to plot a histogram for the price column. Use the hue parameter for the 'cut' column of the diamonds dataset.**

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the diamonds dataset
diamonds = sns.load_dataset("diamonds")

# Plot a histogram for the price column with hue for 'cut'
sns.histplot(data=diamonds, x="price", hue="cut", multiple="stack")
plt.show()
```

**Q5. Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column of the iris dataset.**

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the iris dataset
iris = sns.load_dataset("iris")

# Create a pair plot with hue for 'species'
sns.pairplot(iris, hue="species")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

data = sns.load_dataset("flights")

pivot_data = data.pivot(index="month", columns="year", values="passengers")
sns.heatmap(pivot_data, annot=True, cmap="jet", fmt="d")
plt.title('Heatmap of Number of Passengers by Month and Year')
plt.xlabel('Year')
plt.ylabel('Month')

# Show the plot
plt.show()
```
