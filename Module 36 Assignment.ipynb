Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of Matplotlib.
Matplotlib is a powerful Python library primarily used for creating static, animated, and interactive visualizations. It offers a wide range of plot types to suit diverse data visualization needs.

it is used:

- Versatility: It supports a vast array of plot types, from simple line plots to complex 3D visualizations.
- Customization: Offers extensive control over plot appearance, including colors, markers, labels, and more.
- Integration: Works seamlessly with other scientific Python libraries like NumPy and Pandas.
- Platform independence: Generates plots that can be saved in various formats (PNG, PDF, SVG, etc.).

Five plot types using Pyplot:

- Line plot: Visualizes data points connected by straight lines.
- Scatter plot: Displays data points as individual markers.
- Bar plot: Represents categorical data with rectangular bars.
- Histogram: Visualizes the distribution of numerical data.
- Pie chart: Shows proportions of different categories within a dataset.

Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.


```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(3)

x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
# ... (rest of the code)

# Create the scatter plot
plt.scatter(x, y)

# Add labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot')

# Show the plot
plt.show()
```

A scatter plot is a type of plot that displays values for typically two variables as points on a two-dimensional chart, with the values representing horizontal and vertical axes.

This code generates random data for x and y using NumPy, creates a scatter plot using Matplotlib's scatter function, and adds labels and a title to the plot for better visualization.



Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.

The subplot() function in Matplotlib is used to create multiple plots within a single figure. It allows you to organize and visualize different datasets in a clear and concise manner.

```python
import numpy as np
import matplotlib.pyplot as plt

# Data for four lines
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Create a figure with 2 rows and 2 columns of subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Plot the four lines on the subplots
axs[0, 0].plot(x1, y1, label='Line 1')
axs[0, 0].set_title('Line 1')
axs[0, 0].legend()

axs[0, 1].plot(x2, y2, label='Line 2')
axs[0, 1].set_title('Line 2')
axs[0, 1].legend()

axs[1, 0].plot(x3, y3, label='Line 3')
axs[1, 0].set_title('Line 3')
axs[1, 0].legend()

axs[1, 1].plot(x4, y4, label='Line 4')
axs[1, 1].set_title('Line 4')
axs[1, 1].legend()

# Adjust layout for better spacing
plt.tight_layout()

# Show the plot
plt.show()
```

Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

A bar plot is a graphical representation of categorical data where the height or length of each rectangle corresponds to the value it represents. It's used to compare different categories and visualize their magnitudes.


```python
import numpy as np
import matplotlib.pyplot as plt

company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# Bar plot
plt.bar(company, profit)
plt.xlabel("Company")
plt.ylabel("Profit")
plt.title("Company Profits")
plt.show()

# Horizontal bar plot
plt.barh(company, profit)
plt.xlabel("Profit")
plt.ylabel("Company")
plt.title("Company Profits (Horizontal)")
plt.show()
```

Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

A box plot (or box and whisker plot) is a graphical representation of the distribution of numerical data. It shows the median, quartiles, and outliers of the dataset. It's used to visualize the spread and distribution of data.

```python
import numpy as np
import matplotlib.pyplot as plt

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

data = [box1, box2]

plt.boxplot(data, labels=["Box 1", "Box 2"])
plt.title("Box Plot")
plt.show()
```
