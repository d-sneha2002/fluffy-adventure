**Q1. What are the Probability Mass Function (PMF) and Probability Density Function (PDF)? Explain with an example.**

Probability Mass Function (PMF):
- It defines the probability of a discrete random variable taking on a specific value.
- Used for discrete random variables.
- Example: Probability of getting a specific number when rolling a dice.

Probability Density Function (PDF):
- It defines the probability density for a continuous random variable over a specific range.
- Used for continuous random variables.
- Example: Probability of a person's height falling within a specific range.

**Q2. What is Cumulative Density Function (CDF)? Explain with an example. Why CDF is used?**

Cumulative Density Function (CDF):
- Gives the probability that a random variable is less than or equal to a certain value.   
- Used for both discrete and continuous random variables.
- Example: Probability that a student's height is less than or equal to 170 cm.

CDF is used to:
- Calculate probabilities for ranges of values.
- Compare probability distributions.
- Find percentiles and quantiles.

**Q3. What are some examples of situations where the normal distribution might be used as a model? Explain how the parameters of the normal distribution relate to the shape of the distribution.**

Examples of Normal Distribution:
- Heights of people
- IQ scores
- Measurement errors
- Stock prices

Parameters:
- Mean (µ): Determines the center of the distribution.
- Standard deviation (σ): Controls the spread of the distribution.

A larger standard deviation results in a wider curve, while a smaller standard deviation leads to a narrower curve.

Importance of Normal Distribution:
- Many natural phenomena follow a normal distribution.
- Statistical methods often assume normality.
- Used for modeling and prediction in various fields.

Real-life examples:

**Q5. What is Bernoulli Distribution? Give an Example. What is the difference between Bernoulli Distribution and Binomial Distribution?**

Bernoulli Distribution:
Models a single trial with two possible outcomes (success or failure).   
Example: Flipping a coin once.

Difference:
Bernoulli distribution is for a single trial, while binomial distribution is for multiple trials.

**Q6. Consider a dataset with a mean of 50 and a standard deviation of 10. If we assume that the dataset is normally distributed, what is the probability that a randomly selected observation will be greater than 60? Use the appropriate formula and show your calculations.**

To calculate this, we need to standardize the value using the z-score formula:
- z = (x - μ) / σ
- z = (60 - 50) / 10 = 1

Using a z-table or statistical software, we find the probability of Z > 1 is approximately 0.1587.
Therefore, the probability of a randomly selected observation being greater than 60 is approximately 15.87%.

**Q7. Explain uniform Distribution with an example.**

Uniform Distribution: A probability distribution where all outcomes are equally likely.
- Example: Rolling a fair dice.

**Q8. What is the z score? State the importance of the z score.**

Z-score:
- A standardized value that measures the number of standard deviations a data point is from the mean.
- Important for comparing values from different distributions and for calculating probabilities.

**Q9. What is Central Limit Theorem? State the significance of the Central Limit Theorem.**

Central Limit Theorem (CLT): CLT states that the distribution of sample means approaches a normal distribution as the sample size increases, regardless of the population distribution.
   
Significance:
- Allows us to make inferences about population parameters based on sample statistics.
- Enables the use of normal distribution for statistical tests and confidence intervals.

**Q10. State the assumptions of the Central Limit Theorem.**

- The sample size should be sufficiently large (usually n >= 30).
- The samples should be independently and randomly selected.
- The population distribution should have a finite variance.
